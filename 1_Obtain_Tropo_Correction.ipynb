```python
from pathlib import Path
import concurrent.futures
from tqdm import tqdm
```

```python
job_name = "Marrakesh-1_154_None_0911_noFr" # "Marrakesh-1_154_None_0911"
```

```python
prod_dir = Path(f'products/{job_name}')
assert(prod_dir.exists())
```

```python
gunw_paths = sorted(list(prod_dir.glob('*.nc')))
gunw_paths[:2]
```

Below allows us to utilize the CLI directly within python. The comparable CLI call is `raider.py ++process calcDelaysGUNW -f <gunw_path> -m HRES -interp center_time`

```python
from RAiDER.cli.raider import calcDelaysGUNW

def update_gunw_with_tropo(gunw_path: Path):

    input_arguments = ['--file', str(gunw_path),
                       '--weather-model', 'HRES',
                       '-interp', 'center_time'
                       ]
    calcDelaysGUNW(input_arguments)
    return gunw_path
```

```python
out_paths = list(map(update_gunw_with_tropo, tqdm(gunw_paths[:])))
```
